# Define temporal and spatial templates in Arabic

```python
temporal_templates = ['{entity1} قبل {entity2}', '{entity2} بعد {entity1}', '{entity1} حتى {entity2}',
                      '{entity2} منذ {entity1}', '{entity1} خلال {entity2}', '{entity2} عندما {entity1}']
spatial_templates = ['{entity1} في {entity2}', '{entity1} موجود في {entity2}', '{entity1} يقع في {entity2}']
```

# Load pre-trained FastText embeddings in Arabic



```python

arabic_vectors = KeyedVectors.load_word2vec_format('wiki.ar.vec')
```

# Define a function to extract relations from an Arabic sentence using templates

```python
def extract_relation(sentence, template):
    stop_words = set(nltk.corpus.stopwords.words('arabic'))
    tokens = nltk.word_tokenize(sentence)
    entities = [token for token in tokens if token not in stop_words]
    print(f"entities: {entities}")
    if len(entities) != 2:
        return None
    entity1, entity2 = entities[0], entities[1]
    if template in temporal_templates:
        if entity1 > entity2:
            entity1, entity2 = entity2, entity1
        if template == '{entity1} قبل {entity2}':
            return (entity1, 'قبل', entity2)
        elif template == '{entity2} بعد {entity1}':
            return (entity1, 'بعد', entity2)
        elif template == '{entity1} حتى {entity2}':
            return (entity1, 'حتى', entity2)
        elif template == '{entity2} منذ {entity1}':
            return (entity1, 'منذ', entity2)
        elif template == '{entity1} خلال {entity2}':
            return (entity1, 'خلال', entity2)
        elif template == '{entity2} عندما {entity1}':
            return (entity1, 'عندما', entity2)
    elif template in spatial_templates:
        if entity1 in arabic_vectors.vocab and entity2 in arabic_vectors.vocab:
            similarity_score = arabic_vectors.similarity(entity1, entity2)
            print(f"similarity_score: {similarity_score}")
            if similarity_score > 0.1:
                return (entity1, 'في', entity2)
        return None
    else:
        return None
```

# Define a function to extract all relations from an Arabic text using templates

```python
def extract_relations(text):
    sentences = nltk.sent_tokenize(text)
    print(f"sentences: {sentences}")
    relations = []
    for sentence in sentences:
        for template in temporal_templates + spatial_templates:
            relation = extract_relation(sentence, template)
            if relation is not None:
                relations.append(relation)
                print(f"relation: {relation}")
    return relations
```

# Example usage

```python
text = "المقهى يقع في شارع المدينة. جون ذهب إلى باريس قبل أن يذهب إلى لندن."
relations = extract_relations(text)
print(relations)
```
